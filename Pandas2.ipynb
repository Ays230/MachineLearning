#Missing Data

```python
dfm = pd.read_csv('/content/car-sales-missing-data.csv')
dfm
```

```python
dfm.shape #setir ve sutun sayi
```

```python
dfm.info()
```

```python
dfm.isnull().sum()
```

```python
#bos deyer silmek
dfm.dropna(subset = ['Price'],inplace= True)
dfm
```

```python
dfm.isnull().sum()
```

```python
dfm['Odometer']=dfm['Odometer'].fillna(dfm['Odometer'].mean())
```

```python
dfm
```

```python
dfm['Colour']= dfm['Colour'].fillna(dfm['Colour'].mode()[0])
dfm
```

```python
for col in ['Make','Colour']:
  dfm[col]= dfm[col].fillna(dfm[col].mode()[0])
dfm
```

```python
dfm.isna().sum() #bos deyerimiz qalmadi
```

```python
table_a=pd.DataFrame({'id':['a1','a2','a3','a4'],'Value':['A','B','C','AA']})
table_b=pd.DataFrame({'id':['a1','a2','a3','a5'],'Value':['D','E','F','BB']})
pd.merge(table_a,table_b,on = 'id')
```

```python
table_a=pd.DataFrame({'id':['a1','a2','a3','a4'],'Value':['A','B','C','AA']})
table_b=pd.DataFrame({'id':['a1','a2','a3','a5'],'Value':['D','E','F','BB']})
pd.merge(table_a,table_b,on='id',how='left')
```

```python
table_a=pd.DataFrame({'id1':['a1','a2','a3','a4'],'Value':['A','B','C','AA']})
table_b=pd.DataFrame({'id2':['a1','a2','a3','a5'],'Value':['D','E','F','BB']})
pd.merge(table_a,table_b,left_on ='id1',right_on='id2')
```
